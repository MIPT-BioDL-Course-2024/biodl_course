##Графы. Определение и разновидности

Простыми словами, граф это множество вершин и множество ребер, где каждая вершина может иметь или не иметь ребро с какой-либо иной вершиной графа.

Формально и краткое определение.

Граф - пара множеств G = {E, V} и отображение E → V ✕ V.

V - множество вершин

E - множество ребер e1 = (v1, v2)

Вершины также иногда называют узлами или нодами (nodes). Ребра по английски edges.

#### Adamic-Adar index

Этот индекс довольно хорош для графов с явными кластерами вершин.

Пусть есть граф $G=(V, E)$, с вершинами $u,v \in V$. Обозначим множество соседей вершины $w$ как $\Gamma_w$.

Попробуем придумать некоторую оценку $A(u, v)$ возможности существования ребра $(u, v )$. Какие свойства от нее мы хотим:

- Чем больше общих соседей у вершины -- тем больше эта оценка
- Общие соседи-*хабы* (вершины большой степени) вносят меньший вклад, чем редкие соседи.

Таким свойством обладает индекс Adamic-Adar:

$$ A(u,v) = \sum_{w\in \Gamma_u \cap\Gamma_v}\frac{1}{\log \Gamma_w} $$

Большой индекс можно использовать как сигнал существования ребра.

Вариируя различные параметы вероятностей и длины обхода в глубину и ширину, мы получаем для каждой вершины наборы случайных путей, из которых в последствие может выделить контекст каждой вершины.

Он вводит параметры $p$ и $q$. Во время случайных блужданий вероятности будут пропорциональны:
- Для возвращения к предыдущей вершине -- 1
- Для перехода к вершинам, соседним с предыдущей -- $1/p$
- Для перехода к дальним вершинам -- $1/q$

При $q > p$ мы получаем цепочки, отражающие локальные свойства графа: набор соседей, состав сообществ в графе.

Для $p > q$ получаем больше глобальной информации: связность графа с этой вершиной, центральность вершины.

#### Более математичное описание

Изначально нужно в узлах оставить какие-то данные, получив какую матрицу данных узлов $X$, размером $N_{nodes}\times E$.

К этому набору эмбеддингов применяется обучаемая матрица $\theta$ размером ($E\times E_{new}$).

Новый эмбеддинг узла -- среднее всех репрезентаций его соседей, пропущенные через активационный слой $\sigma$.

В матричном виде это можно записать так:

$$X'=\sigma(D^{-0.5}AD^{-0.5}X\theta)$$

- $A$ -- матрица смежности графа
- $D$ -- диагональная матрица степеней вершин

Получаемый слой называется *GCN*.

К этим эмбеддингам можно всегда прикрутить pooling для получения репрезентаций целых графов.

По сути любая графовая свертка может быть выражена формулой:

$$\mathbf{x}_i^{(k)} = \gamma^{(k)} \left( \mathbf{x}_i^{(k-1)}, \bigoplus_{j \in \mathcal{N}(i)} \, \phi^{(k)}\left(\mathbf{x}_i^{(k-1)}, \mathbf{x}_j^{(k-1)},\mathbf{e}_{j,i}\right) \right),$$

где:
- $\mathbf{x}_i^{(k)}$ -- вектор в узле $i$ после $k$ сверток
- $\mathbf{e}_{j,i}$ -- вектор ребра между i, j
- $\gamma^{(k)}$ и $\phi^{(k)}$ -- дифференцируемые функции $k$-той свертки
- $\bigoplus$ -- операция аггрегации, инвариантная к перестановке аргументов

### GAT

GAT (graph attention network) --пожалуй, одна из наиболее популярных сверток. Она использует механизм attention.

Оператор GATv2 из статьи ["How Attentive are Graph Attention Networks?"](https://arxiv.org/abs/2105.14491):

$$\mathbf{x}^{\prime}_i = \alpha_{i,i}\mathbf{\Theta}_{s}\mathbf{x}_{i} +
\sum_{j \in \mathcal{N}(i)}
\alpha_{i,j}\mathbf{\Theta}_{t}\mathbf{x}_{j},$$

$$\alpha_{i,j} =
\frac{
\exp\left(\mathbf{a}^{\top}\mathrm{LeakyReLU}\left(
\mathbf{\Theta}_{s} \mathbf{x}_i + \mathbf{\Theta}_{t} \mathbf{x}_j
\right)\right)}
{\sum_{k \in \mathcal{N}(i) \cup \{ i \}}
\exp\left(\mathbf{a}^{\top}\mathrm{LeakyReLU}\left(
\mathbf{\Theta}_{s} \mathbf{x}_i + \mathbf{\Theta}_{t} \mathbf{x}_k
\right)\right)}.$$

В случае, если есть аттрибуты ребер attention меняется так:

$$\alpha_{i,j} =
\frac{
\exp\left(\mathbf{a}^{\top}\mathrm{LeakyReLU}\left(
\mathbf{\Theta}_{s} \mathbf{x}_i
+ \mathbf{\Theta}_{t} \mathbf{x}_j
+ \mathbf{\Theta}_{e} \mathbf{e}_{i,j}
\right)\right)}
{\sum_{k \in \mathcal{N}(i) \cup \{ i \}}
\exp\left(\mathbf{a}^{\top}\mathrm{LeakyReLU}\left(
\mathbf{\Theta}_{s} \mathbf{x}_i
+ \mathbf{\Theta}_{t} \mathbf{x}_k
+ \mathbf{\Theta}_{e} \mathbf{e}_{i,k}]
\right)\right)}.$$

```python
# Install required packages.
!pip install -q torch-scatter -f https://pytorch-geometric.com/whl/torch-1.10.0+cu113.html
!pip install -q torch-sparse -f https://pytorch-geometric.com/whl/torch-1.10.0+cu113.html
!pip install -q git+https://github.com/rusty1s/pytorch_geometric.git

import torch
from torch_geometric.datasets import TUDataset

dataset = TUDataset(root='data/TUDataset', name='MUTAG')

print()
print(f'Dataset: {dataset}:')
print('====================')
print(f'Number of graphs: {len(dataset)}')
print(f'Number of features: {dataset.num_features}')
print(f'Number of classes: {dataset.num_classes}')

data = dataset[0]  # Get the first graph object.

print()
print(data)
print('=============================================================')

# Gather some statistics about the first graph.
print(f'Number of nodes: {data.num_nodes}')
print(f'Number of edges: {data.num_edges}')
print(f'Average node degree: {data.num_edges / data.num_nodes:.2f}')
print(f'Has isolated nodes: {data.has_isolated_nodes()}')
print(f'Has self-loops: {data.has_self_loops()}')
print(f'Is undirected: {data.is_undirected()}')
```

```python
torch.manual_seed(12345)
dataset = dataset.shuffle()

train_dataset = dataset[:150]
test_dataset = dataset[150:]

print(f'Number of training graphs: {len(train_dataset)}')
print(f'Number of test graphs: {len(test_dataset)}')
```

```python
from torch_geometric.loader import DataLoader

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)

for step, data in enumerate(train_loader):
    print(f'Step {step + 1}:')
    print('=======')
    print(f'Number of graphs in the current batch: {data.num_graphs}')
    print(data)
    print()
```

```python

from torch.nn import Linear
import torch.nn.functional as F
from torch_geometric.nn import GCNConv
from torch_geometric.nn import global_mean_pool


class GCN(torch.nn.Module):
    def __init__(self, hidden_channels):
        super(GCN, self).__init__()
        torch.manual_seed(12345)
        self.conv1 = GCNConv(dataset.num_node_features, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, hidden_channels)
        self.conv3 = GCNConv(hidden_channels, hidden_channels)
        self.lin = Linear(hidden_channels, dataset.num_classes)

    def forwa
rd(self, x, edge_index, batch):
        # 1. Obtain node embeddings
        x = self.conv1(x, edge_index)
        x = x.relu()
        x = self.conv2(x, edge_index)
        x = x.relu()
        x = self.conv3(x, edge_index)

        # 2. Readout layer
        x = global_mean_pool(x, batch)  # [batch_size, hidden_channels]

        # 3. Apply a final classifier
        x = F.dropout(x, p=0.5, training=self.training)
        x = self.lin(x)

        return x

model = GCN(hidden_channels=64)
print(model)
```

```python
from IPython.display import Javascript
display(Javascript('''google.colab.output.setIframeHeight(0, true, {maxHeight: 300})'''))

model = GCN(hidden_channels=64)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
criterion = torch.nn.CrossEntropyLoss()

def train():
    model.train()

    for data in train_loader:  # Iterate in batches over the training dataset.
         out = model(data.x, data.edge_index, data.batch)  # Perform a single forward pass.
         loss = criterion(out, data.y)  # Compute the loss.
         loss.backward()  # Derive gradients.
         optimizer.step()  # Update parameters based on gradients.
         optimizer.zero_grad()  # Clear gradients.

def test(loader):
     model.eval()

     correct = 0
     for data in loader:  # Iterate in batches over the training/test dataset.
         out = model(data.x, data.edge_index, data.batch)
         pred = out.argmax(dim=1)  # Use the class with highest probability.
         correct += int((pred == data.y).sum())  # Check against ground-truth labels.
     return correct / len(loader.dataset)  # Derive ratio of correct predictions.


for epoch in range(1, 171):
    train()
    train_acc = test(train_loader)
    test_acc = test(test_loader)
    print(f'Epoch: {epoch:03d}, Train Acc: {train_acc:.4f}, Test Acc: {test_acc:.4f}')
```
